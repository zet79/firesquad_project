```python
import matplotlib.pyplot as plt
from pathlib import Path
import os
from tensorflow.keras.preprocessing.image import load_img
import random
```

```python
def show_images(directory, num_images=5):
    total = 0
    category_counts = []
    categories = os.listdir(directory)

    for category in categories:
        category_path = os.path.join(directory, category)
        images = os.listdir(category_path)[:num_images]
        print(f"Category: {category}, Total images: {len(os.listdir(category_path))}")
        
        total += len(os.listdir(category_path))
        category_counts.append(len(os.listdir(category_path)))

        plt.figure(figsize=(15, 5))
        for i, img in enumerate(images):
            img_path = os.path.join(category_path, img)
            loaded_img = load_img(img_path, target_size=(224, 224))
            plt.subplot(1, num_images, i + 1)
            plt.imshow(loaded_img)
            plt.axis('off')
            plt.title(category)
        plt.show()

    no_fire_percentage = (category_counts[0] / total) * 100
    fire_percentage = (category_counts[1] / total) * 100
    print(f"\nPercentage of 'No_Fire': {no_fire_percentage:.2f}%")
    print(f"Percentage of 'Fire': {fire_percentage:.2f}%\n\n")

# Adjust path depending on current script location
project_root = Path().resolve().parent
flame_data_path = project_root / "data" / "FLAME"

print("Training images:")
show_images(os.path.join(flame_data_path, "Training"))

print("Test images:")
show_images(os.path.join(flame_data_path, "Test"))
```

```python
def verify_size(directory):
    categories = os.listdir(directory)
    for category in categories:
        category_path = os.path.join(directory, category)
        images = os.listdir(category_path)
        for img in images[:5]:  # Check 5 images per category
            img_path = os.path.join(category_path, img)
            loaded_img = load_img(img_path)
            print(f"Category: {category}, Image: {img}, Size: {loaded_img.size}, Format: {img.split('.')[-1]}")

print("Verifying image size and format:")
verify_size(os.path.join(flame_data_path, "Training"))
```

```python
def random_show(directory):
    categories = os.listdir(directory)
    category = random.choice(categories)
    category_path = os.path.join(directory, category)
    img = random.choice(os.listdir(category_path))
    img_path = os.path.join(category_path, img)
    loaded_img = load_img(img_path, target_size=(224, 224))
    plt.imshow(loaded_img)
    plt.axis('off')
    plt.title(f"Category: {category}")
    plt.show()

print("Random image:")
random_show(os.path.join(flame_data_path, "Training"))
```
